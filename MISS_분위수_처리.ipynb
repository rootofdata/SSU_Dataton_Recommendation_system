도서대출분위수

```python
#대출 도수 데이터프레임
bookloan_freq_book = pd.DataFrame(bookloan['서명'].value_counts().reset_index())
bookloan_freq_book.columns = ['서명', '도수']
```

```python
bookloan_quan = bookloan_freq_book.copy()
```

```python
#대출 도수 5분위수 지정
loan_q1 = bookloan_freq_book['도수'].quantile(.2)
loan_q2 = bookloan_freq_book['도수'].quantile(.4)
loan_q3 = bookloan_freq_book['도수'].quantile(.6)
loan_q4 = bookloan_freq_book['도수'].quantile(.8)
```

```python
#대출 도수 분위수화
bookloan_quan.columns = ['서명','대출분위수']

for i in range(len(bookloan_quan)):
    if bookloan_quan['대출분위수'][i] > loan_q4 :
        bookloan_quan['대출분위수'][i] = 5
    elif bookloan_quan['대출분위수'][i] > loan_q3 :
        bookloan_quan['대출분위수'][i] = 4
    elif bookloan_quan['대출분위수'][i] > loan_q2 :
        bookloan_quan['대출분위수'][i] = 3
    elif bookloan_quan['대출분위수'][i] > loan_q1 :
        bookloan_quan['대출분위수'][i] = 2
    else :
        bookloan_quan['대출분위수'][i]=1
```

소속별 분위수

```python
#소속별 대출 도수표 '소속명+도수표'로 저장
major_list = bookloan['소속'].unique()
for i in major_list:
    globals()[str(i)+'loan'] = bookloan[bookloan['소속'] == i]
    globals()[str(i)+'도수표'] = pd.DataFrame(globals()[str(i)+'loan']['서명'].value_counts().reset_index())
    globals()[str(i)+'도수표'].columns = ['서명','도수']
```

```python
#소속별 대출도수표 -> 소속별 대출 분위수로 변경
qu = [4,3,2,1]
for i in major_list:
    globals()[str(i)+'분위수'] = globals()[str(i)+'도수표'].copy()
    globals()[str(i)+'분위수'].columns = ['서명','분위수']
    
    #소속별 5분위수 지정
    for s in qu :
        globals()[str(i)+'q'+str(s)] = globals()[str(i)+'분위수']['분위수'].quantile(0.2*s)
    
    #소속별 도출 도수 분위수화
    for t in range(len(globals()[str(i)+'분위수'])) :
        if globals()[str(i)+'분위수']['분위수'][t] > globals()[str(i)+'q'+str(qu[0])]:
            globals()[str(i)+'분위수']['분위수'][t] = 5 
        elif globals()[str(i)+'분위수']['분위수'][t] > globals()[str(i)+'q'+str(qu[1])]:
            globals()[str(i)+'분위수']['분위수'][t] = 4 
        elif globals()[str(i)+'분위수']['분위수'][t] > globals()[str(i)+'q'+str(qu[2])]:
            globals()[str(i)+'분위수']['분위수'][t] = 3
        elif globals()[str(i)+'분위수']['분위수'][t] > globals()[str(i)+'q'+str(qu[3])]:
            globals()[str(i)+'분위수']['분위수'][t] = 2
        else :
            globals()[str(i)+'분위수']['분위수'][t] = 1
```
